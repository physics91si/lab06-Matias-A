# Part 1

```python
# TODO define gaussian function here
def gaussian(x, mean, std):
    return (1/(std*np.sqrt(2*np.pi)))*np.exp(-(1/2)*np.square((x-mean)/std))
    
```

```python
# TODO plot gaussians here
x=np.linspace(-5,5,100)
y=gaussian(x,0,2)
plt.plot(x,y)
```

```python
# TODO fit gaussians, with known parameters, here

# TODO plot result

params, _= scipy.optimize.curve_fit(gaussian, x, y)
plt.plot(x,y)
plt.plot(x,gaussian(x, *params), linestyle='dashed')
```

```python
# TODO: fit a function to this data and plot it
def gaussian(x, mean, std, amp):
    return amp*np.exp(-(1/2)*np.square((x-mean)/std))

def doublegaussian(x, mean1, std1, amp1, mean2, std2, amp2):
    gaussian1=gaussian(x,mean1,std1, amp1)
    gaussian2=gaussian(x,mean2,std2, amp1)
    return gaussian1+gaussian2
params2, _= scipy.optimize.curve_fit(doublegaussian, datax,-datay, p0=(13128,1,0.000025,13132,1.1,0.000025))
print(params2)
plt.plot(datax, doublegaussian(datax, *params2))
plt.plot(datax,-datay)
```

```python
def flux_for_dipole(x_pos, y_pos, Q_vals, X, Y):
    """
    Returns the flux based on the information of the charges
    and the grid of the plot.
    Code modified from Physics 113, written by Blas Cabrera
    """
    flux = np.zeros((len(X), len(Y)))
    x_m = np.outer(np.ones((X.size)), X); 
    y_m = np.outer(Y, np.ones((Y.size))); 
    for n in range(len(Q_vals)):
        flux += 2*np.pi*Q_vals[n]*(1 - (x_pos[n]-x_m)/np.sqrt((y_pos[n]-y_m)**2+(x_pos[n]-x_m)**2))
    return flux
```

```python
# TODO create arrays of information
xpos = np.asarray((-10,22))
ypos = np.asarray((6,-3))
q = (-1,1)
X=np.linspace(-31.5,31.5,64)
Y=np.linspace(-31.5,31.5,64)
# TODO get flux
flux = flux_for_dipole(xpos, ypos, q, X, Y)

# TODO plot contours of the flux
plt.contour(X,Y,flux,20)
plt.plot(xpos, ypos, "o", ms='15', c='black')
```
